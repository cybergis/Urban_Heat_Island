```python
import miceforest as mf
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data_format/data0.csv")
for i in range(1, 10):
    filename = "./data_format/data"+str(i)+".csv"
    curr = pd.read_csv(filename)
    data = data.append(curr, ignore_index=True)
```

```python
data
```

```python
df = data.drop('Unnamed: 0', 1)
df
```

```python
df["partical"] = 0
df
```

```python
curr_mice = mf.KernelDataSet(
  df,
  save_all_iterations=True,
  random_state=1991
)
```

```python
curr_mice.mice(3)
```

```python
completed_data = curr_mice.complete_data()
```

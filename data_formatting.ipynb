```python
## Process & Interpolate for node_with_band.csv
## Dictionary d[(node_id, datetime,band_id)] = (lat, lon, value)
```

```python
import datetime
import pandas as pd
from scipy import interpolate
import numpy as np
```

```python
node_band = pd.read_csv("node_with_band.csv")
node_band
```

```python
d = {}
for index, row in node_band.iterrows():
    ls_607 = [row["607_1"],row["607_2"],row["607_3"],row["607_4"],row["607_5"],row["607_6"],row["607_7"]]
    ls_725 = [row["725_1"],row["725_2"],row["725_3"],row["725_4"],row["725_5"],row["725_6"],row["725_7"]]
    ls_911 = [row["911_1"],row["911_2"],row["911_3"],row["911_4"],row["911_5"],row["911_6"],row["911_7"]]
    
    s= datetime.datetime(2018, 6, 21)
    
    x_607 = datetime.datetime(2018, 6, 7)
    x_725 = datetime.datetime(2018, 7, 25)
    x_911 = datetime.datetime(2018, 9, 11)
    ## Fit for each band
    for i in range(0, 7):
        x = np.array([(x_607-s).days, (x_725-s).days, (x_911-s).days])
        y = np.array([ls_607[i], ls_725[i], ls_911[i]])
        ## polynomial Interpolation for each band
        f = interpolate.interp1d(x, y, fill_value="extrapolate")
        
        for dot in range(0, 95):
            curr_date = s+datetime.timedelta(days=dot)
            curr_value = f(dot)
            curr_band = i+1
            curr_id = row["node_id"]
            d[(curr_id, curr_date, curr_band)]=float(curr_value)
```

```python
## Use d to map current data
## id, minutes of days, days of time, year, humidity, PM25, SO2, particle 10, light intensity, lat, lon, band1-7, temperature
```

```python
import numpy as np
def remove_outlier_and_get_value(dt):
    ## remove outlier with 1.5IQR
    if (len(dt)==0):
        return "NA"
    Q1 = np.quantile(dt,0.25)
    Q2 = np.quantile(dt,0.5)
    Q3 = np.quantile(dt,0.75)
    intr_qr = Q3-Q1
    max_v = Q3+1.5*intr_qr
    min_v = Q1-1.5*intr_qr
    curr_list = []
    for i in range(0, len(dt)):
        if (dt[i]<=max_v and dt[i]>=min_v):
            ## not a outlier
            curr_list.append(dt[i])
    return np.median(curr_list)
```

```python
data = pd.read_csv("data.csv")
data
```

```python
sample_list = set(list(zip(data["timestamp"].values.tolist(),data["node_id"].values.tolist())))
```

```python
df = pd.DataFrame(columns=["temp", "humidity", "PM25", "SO2", "partical", "light", "band1", "band2", "band3", "band4","band5","band6","band7","dist_to_cent", "tod", "dss", "year"])

for (curr_time, curr_node) in sample_list:
    if (df.shape[0]%500==0):
        print(df.shape[0])
    curr_df = data[(data["timestamp"]==curr_time) & (data["node_id"]==curr_node)]
    ## temperature
    temp = remove_outlier_and_get_value(curr_df[curr_df["parameter"]=="temperature"]["value_hrf_average"].values.tolist())
    ## humidity
    humidity =  remove_outlier_and_get_value(curr_df[curr_df["parameter"]=="humidity"]["value_hrf_average"].values.tolist())
    ## PM2.5
    PM25 = remove_outlier_and_get_value(curr_df[curr_df["parameter"]=="pm2_5"]["value_hrf_average"].tolist()+curr_df[curr_df["parameter"]=="pm25_cf"]["value_hrf_average"].tolist())
    ## SO2
    SO2 = remove_outlier_and_get_value(curr_df[curr_df["sensor"]=="so2"]["value_hrf_average"].values.tolist())
    ## 10um particle
    partical = remove_outlier_and_get_value(curr_df[curr_df["parameter"]=="10um_particle"]["value_hrf_average"].values.tolist())
    ## light intensity
    light_intensity_df = curr_df[(curr_df["subsystem"]=="lightsense") & (curr_df["parameter"]=="intensity") & ((curr_df["sensor"]=="tsl260rd") | (curr_df["sensor"]=="tsl250rd"))]
    light_intensity = remove_outlier_and_get_value(light_intensity_df[light_intensity_df["value_hrf_average"]<=124]["value_hrf_average"].values.tolist())
    ## tod  time of day = minutes since 00:00
    hm = curr_time.split(" ")[1].split(":")
    tod = int(hm[0])*60+int(hm[1])
    ## dss day since start = day since start 6/21
    dmy = curr_time.split(" ")[0].split("/")
    dss_v = datetime.datetime(year = int(dmy[0]), month = int(dmy[1]), day = int(dmy[2])) - datetime.datetime(year = 2018, month = 6, day = 21)
    dss = dss_v.days
    ## distance to centroid
    cenlat = 41.878689
    cenlon = -87.640329
    lat = node_band[node_band["node_id"]==curr_node]["lat"].values[0]
    lon = node_band[node_band["node_id"]==curr_node]["lon"].values[0]
    dist_to_cent = np.sqrt((cenlat-lat)*(cenlat-lat)+(cenlon-lon)*(cenlon-lon))
    ## Band value

    curr_d = datetime.datetime(year = int(dmy[0]), month = int(dmy[1]), day = int(dmy[2]))
    band1 = d[curr_node, curr_d, 1]
    band2 = d[curr_node, curr_d, 2]
    band3 = d[curr_node, curr_d, 3]
    band4 = d[curr_node, curr_d, 4]
    band5 = d[curr_node, curr_d, 5]
    band6 = d[curr_node, curr_d, 6]
    band7 = d[curr_node, curr_d, 7]

    if (temp!="NA"):
        df.loc[len(df)]=[temp, humidity, PM25, SO2, partical, light_intensity, band1, band2, band3, band4, band5, band6, band7, dist_to_cent, tod, dss, 2018] 

df
```
